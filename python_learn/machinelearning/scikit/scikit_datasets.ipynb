```python
# 自定义数据集
# 生成聚类数据集
#datasets.make_blobs(n_samples=100, n_features=2, centers=3, cluster_std=1.0, center_box=(-10.0, 10.0), shuffle=True, random_state=None)
#其中，n_samples是待生成的样本的总数，n_features是每个样本的特征数，centers表示类别数，cluster_std表示每个类别的方差，center_box表示样本中心的范围，shuffle是否将数据打乱，random_state用于随机生成数据。
from sklearn import datasets
import matplotlib.pyplot as plt
X, y = datasets.make_blobs(n_samples=100, n_features=2, centers=3, cluster_std=1.0, center_box=(-10.0, 10.0), shuffle=True, random_state=None)

plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
#生成同心圆样本点
#datasets.make_circles(n_samples=100, shuffle=True, noise=None, random_state=None, factor=0.8)
#其中，n_samples是待生成的样本的总数，shuffle是否将数据打乱，noise表示样本点的噪声，factor表示内外圆接近程度，越大越接近，上限为1。
from sklearn import datasets
import matplotlib.pyplot as plt
X, y = datasets.make_circles(n_samples=1000, shuffle=True, noise=0.1, random_state=None, factor=0.8)

plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
#生成模拟分类数据集
#datasets.make_classification(n_samples=100, n_features=20, n_informative=2, n_redundant=2, n_repeated=0, n_classes=2, n_clusters_per_class=2, weights=None, flip_y=0.01, class_sep=1.0, hypercube=True, shift=0.0, scale=1.0, shuffle=True, random_state=None)
#其中，n_samples是待生成的样本的总数，n_features是每个样本的特征数，n_informative是用于分类的特征数，n_redundant是冗余特征数，n_repeated是重复特征数，n_classes是输出的类别数，weights是每个类别的权重，flip_y是噪声，class_sep是用于随机超平面的距离，hypercube表示是否绘制超立方体，shift是超立方体中心的偏移，scale是超立方体的尺寸，shuffle是否将数据打乱，random_state用于随机生成数据。
from sklearn import datasets
import matplotlib.pyplot as plt
X, y = datasets.make_classification(n_samples=1000, n_features=20, n_informative=2, n_redundant=2, n_repeated=0, n_classes=2, n_clusters_per_class=2, weights=None, flip_y=0.01, class_sep=1.0, hypercube=True, shift=0.0, scale=1.0, shuffle=True, random_state=None)
print(X.shape)
print(y.shape)
```

```python
#生成太极型非凸集样本点
#datasets.make_moons(n_samples=100, shuffle=True, noise=None, random_state=None)
#其中，n_samples是待生成的样本的总数，shuffle是否将数据打乱，noise表示样本点的噪声。
from sklearn import datasets
import matplotlib.pyplot as plt
X, y = datasets.make_moons(n_samples=800, shuffle=True, noise=0.1, random_state=None)

plt.scatter(X[:, 0], X[:, 1], c=y)
```
